# The Perception Step

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Reading in Color Images
To read in an image using Python you can use the opencv and to display matplotlib

```python
# Reading image
IMAGE_PATH='images/sample.jpg'
img_brg = cv2.imread(IMAGE_PATH)

# Converting BGR to RGB
img = cv2.cvtColor(img_brg, cv2.COLOR_BGR2RGB)

plt.imshow(img)
print("Image Shape: ", img.shape)
```

### Coverting image in resptective color channel

```python
# Making three copy of image
red_channel = np.copy(img)
green_channel = np.copy(img)
blue_channel = np.copy(img)

red_channel[:,:,[1,2]] = 0
green_channel[:,:,[0,2]] = 0
blue_channel[:,:,[0,1]] = 0


# Display sobel -1 depth
fx, plots = plt.subplots(1, 3, figsize=(20,10))

plots[0].set_title("Red Channel")
plots[0].imshow(red_channel)

plots[1].set_title("Green Channel")
plots[1].imshow(green_channel)

plots[2].set_title("Blue Channel")
plots[2].imshow(blue_channel)
```

### Color Thresholding

```python
# Define a function to perform a color threshold
def color_thresh(img, rgb_thresh=(0, 0, 0)):
    # Create an empty array the same size in x and y as the image 
    # but just a single channel
    binary_image = np.zeros_like(img[:,:,0])
    
    # Apply the thresholds for RGB and 
    # assign 1's where threshold was exceeded
    above_thresh = (img[:,:,0] > rgb_thresh[0]) & (img[:,:,1] > rgb_thresh[1]) & (img[:,:,2] > rgb_thresh[2])
    binary_image[above_thresh] = 1
    
    return binary_image
```

```python
red_threshold = 128
green_threshold = 128
blue_threshold = 120
rgb_threshold = (red_threshold, green_threshold, blue_threshold)

binary_image = color_thresh(img, rgb_threshold)

plt.imshow(binary_image, cmap='gray')
```
